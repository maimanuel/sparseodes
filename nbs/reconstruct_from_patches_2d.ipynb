```python
from sklearn.feature_extraction.image import extract_patches_2d
from sklearn.feature_extraction.image import reconstruct_from_patches_2d
import numpy as np
import pylab as plb
%pylab inline
```

```python
#original image
im = np.arange(17**2).reshape(17,17)
im_size = im.shape

# patch size list
ps_list = [2,4,8,12,16]

#show original and reconstructions
figure(figsize = (10,2))
subplot(1,1+len(ps_list),1)
imshow(im)

title('original')
for i,ps in enumerate(ps_list):
    patch_size =(ps,ps) 
    #turn image into patches
    patches = extract_patches_2d(im,patch_size)
    #turn patches into image
    im_new = reconstruct_from_patches_2d(patches,im_size)
    
    subplot(1,1+len(ps_list),2+i)
    imshow(im_new)
    title('rec w ps=' + str(patch_size),size = '10')

tight_layout()    
```
